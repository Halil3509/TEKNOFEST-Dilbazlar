```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from transformers_interpret import MultiLabelClassificationExplainer

model_name = "halilibr/dilbazlar-bert-uncased-anxiety-disorders-recognition-balanced-tr-model-acc-92.7"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)


cls_explainer = MultiLabelClassificationExplainer(model, tokenizer)
```

```python
word_attributions = cls_explainer("5 yıldır evden çıkmıyorum. çıkmak istiyorum ama çok korkuyorum.")
```

```python
cls_explainer.visualize("multilabel_viz.html")
```
