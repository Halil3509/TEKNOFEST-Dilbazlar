# Model için veri artırımı talimatı
instruction = """
Please produce similar Turkish sentences using the sentence below. Produce 9 similar sentences, keeping the meaning of
each sentence but changing the language structure and word choices. The sentences should contain symptoms of both depression and anxiety. 
The sentences should be natural, varied and similar in meaning. Just create sentences, no comments or anything. The result must be in JSON format as following:

JSON format:
        {"Sentence": <the produced sentence>}

as an example for you:

Input sentence: "Bugün hiç bitmeyecekmiş gibi."

Output JSON:
{"Sentence":"Bu günün de geçmeyeceği gibi bir his var içimde.",
"Sentence":"Gün geçmiyor gibi hissediyorum, bu gün de öyle görünüyor.",
"Sentence":"Bu günün hiç bitmeyecek gibi olduğu hissine kapıldım.",
"Sentence":"Zamanın geçmediği bir gün daha gibi geliyor.",
"Sentence":"Bu gün de her şey çok yavaş ilerliyor, sanki hiç geçmeyecekmiş gibi."]
"""

# Batch işlemi için fonksiyon
def process_batch(sentences, batch_size):
    all_augmented_samples = []
    for i in tqdm(range(0, len(sentences), batch_size)):
        batch = sentences[i:i + batch_size]
        batch_results = []
        for sentence in batch:
            try:
                response = model.generate_content(f"{instruction}\n\nCümle: \"{sentence}\"")
                batch_results.append(response.text.strip())
                time.sleep(4.1)  # API çağrıları arasında bekleme süresi
            except Exception as e:
                print(f"Bir hata oluştu: {e}")
                batch_results.append(None)  # Hata durumunda None ekleyin
        all_augmented_samples.extend(batch_results)
        # Batch sonuçlarını kaydedin
        with open(f"augmented_batch_{i // batch_size + 1}.json", "w", encoding='utf-8') as f:
            json.dump(batch_results, f, ensure_ascii=False, indent=4)
    return all_augmented_samples

# Batch büyüklüğünü belirleyin
batch_size = 60

# Batch işlemi
augmented_data = process_batch(sentences, batch_size)

# Toplam artırılmış veri sayısını yazdırın
print(f"Toplam artırılmış veri sayısı: {len(augmented_data)}")


```python
import json
import glob

json_folder_path = 'Data/depression_anxiety_multilabelled_data/'

# Tüm JSON dosyalarının yolunu al
json_files = glob.glob(f"{json_folder_path}/*.json")

# Tüm verileri toplayacak liste
augmented_data = []

# JSON dosyalarını oku ve verileri listeye ekle
for json_file in json_files:
    with open(json_file, 'r', encoding='utf-8') as f:
        # JSON dosyasının içeriğini yükle
        data = json.load(f)
        # Listeye ekle
        augmented_data.extend(data)

# Tüm verileri yazdırın veya işleyin
print(f"Toplam veri sayısı: {len(augmented_data)}")
```

```python
augmented_data = [item for item in augmented_data if item is not None]
```

```python
import re
pattern = re.compile(r':\s*"(.*?)"', re.DOTALL)
augmented_final_data = []

for sentence in augmented_data:
    matches = pattern.findall(sentence)
    for match in matches:
        augmented_final_data.append(match)
```
